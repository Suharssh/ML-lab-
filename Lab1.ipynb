```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0, 10.0)

df = pd.read_csv("C:/Users/student/Downloads/headbrain.csv")
print(df.shape)
df.head()
```

```python
X = df['Head Size(cm^3)'].values
Y = df['Brain Weight(grams)'].values
```

```python
mean_X = np.mean(X)
mean_Y = np.mean(Y)

n = len(X)

numer = 0
denom = 0

for i in range(n):
    numer+=(X[i]-mean_X)*(Y[i]-mean_Y) #finding the sum of residues
    denom+=(X[i]-mean_X)**2
    
b1 = numer/denom
b0 = mean_Y + b1*(mean_X)

print(b0,b1) #regression coefficients
```

```python
max_x = np.max(X)+100
min_x = np.min(X)-100

x = np.linspace(min_x,max_x,1000)
y = b0+b1*x
plt.plot(x,y,color = 'red',label = 'Regression line')
plt.scatter(X, Y, c='green', label='Data points')
plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```
